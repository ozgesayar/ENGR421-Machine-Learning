```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.spatial as spa
```

#### Import Data 

```python
# read data into memory
data_set = np.genfromtxt("hw09_data_set.csv", delimiter = ",")

x = data_set[:,0]
y = data_set[:,1]

# get number of samples and number of features
N = len(x)
K = 9
```

#### Parameters 

```python
#mean parameters
cluster_means = np.array([[+5.0, +5.0], #mean1
                         [-5.0, +5.0], #mean2
                         [-5.0, -5.0], #mean3
                         [+5.0, -5.0], #mean4
                         [+5.0, +0.0], #mean5
                         [+0.0, +5.0], #mean6
                         [-5.0, +0.0], #mean7
                         [+0.0, -5.0], #mean8
                         [+0.0, +0.0]]) #mean9
#covarience parameters
cluster_covariences = np.array([[[+0.8, -0.6],  #covarience1
                                 [-0.6, +0.8]],
                                [[+0.8, +0.6],  #covarience2
                                 [+0.6, +0.8]],
                                [[+0.8, -0.6],  #covarience3
                                 [-0.6, +0.8]],
                                [[+0.8, +0.6],  #covarience4
                                 [+0.6, +0.8]],
                                [[+0.2, +0.0],  #covarience5
                                 [+0.0, +1.2]],
                                [[+1.2, +0.0],  #covarience6
                                 [+0.0, +0.2]],
                                [[+0.2, +0.0],  #covarience7
                                 [+0.0, +1.2]],
                                [[+1.2, +0.0],  #covarience8
                                 [+0.0, +0.2]],
                                [[+1.6, +0.0],  #covarience9
                                 [+0.0, +1.6]]])

cluster_sizes = np.array([100, 100, 100, 100, 100, 100, 100, 100, 200])
```

#### Plotting Data 

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y, "b.", markersize=10)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.show()
```

#### Euclidean Matrix 

```python
dist = np.zeros((N,N))
b = np.zeros((N,N))
for i in range (N):
    for j in range (N):
        dist[i][j] = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2)
        if i != j:
            if dist[i][j] < 2.0:
                b[i][j] = 1
            else:
                b[i][j] = 0
            
b = np.array(b) 
b = b.astype(int)
print(b)
```

```python
plt.figure(figsize=(6, 6))

for i in range(N):
    for j in range(i, N):
        if b[i, j]:
            x1 = [data_set[i,0], data_set[j,0]]
            x2 = [data_set[i,1], data_set[j,1]]
            plt.plot(x1, x2, "-", color="gray")

plt.plot(x, y, "k.", markersize=12)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.show()
```

#### D and L matrices 

```python
#D matrix
d_matrix = np.zeros((N,N))
one_sum = 0
for i in range (N):
    one_sum = 0
    for j in range (N):
        one_sum = one_sum + b[i][j] 
    d_matrix[i][i]= one_sum

d_matrix = d_matrix.astype(int)
```

```python
#L matrix
d_sqrt = np.sqrt(np.linalg.inv(d_matrix))
matrix = d_matrix - b
l_symmetric = np.dot(d_sqrt, np.dot(matrix, d_sqrt))

print(l_symmetric[0:5 , 0:5])
```

#### Eigenvectors 

```python
R = 5

eigenvalues, eigenvectors = np.linalg.eig(l_symmetric)
eigenvalues = np.real(eigenvalues)
eigenvectors = np.real(eigenvectors)

Z = eigenvectors[:, np.argsort(eigenvalues)[1:(R+1)]]

print(Z[0:5, 0:5])
```

#### Running K means clustering algorithm

```python
def update_centroids(memberships, X):
    if memberships is None:
        # initialize centroids
        centroids = X[np.random.choice(range(N), K, False),:]
    else:
        # update centroids
        centroids = np.vstack([np.mean(X[memberships == k,:], axis = 0) for k in range(K)])
    return(centroids)

def update_memberships(centroids, X):
    # calculate distances between centroids and data points
    D = spa.distance_matrix(centroids, X)
    # find the nearest centroid for each data point
    memberships = np.argmin(D, axis = 0)
    return(memberships)

def plot_current_state(centroids, memberships, X):
    cluster_colors = np.array(["#1f78b4", "#33a02c", "#e31a1c", "#ff7f00", "#6a3d9a", "#b15928",
                               "#a6cee3", "#b2df8a", "#fb9a99", "#fdbf6f", "#cab2d6", "#ffff99"])
    if memberships is None:
        plt.plot(X[:,0], X[:,1], ".", markersize = 10, color = "black")
    else:
        for c in range(K):
            plt.plot(X[memberships == c, 0], X[memberships == c, 1], ".", markersize = 10,
                     color = cluster_colors[c])
    for c in range(K):
        plt.plot(centroids[c, 0], centroids[c, 1], "s", markersize = 12,
                 markerfacecolor = cluster_colors[c], markeredgecolor = "black")
    plt.xlabel("x1")
    plt.ylabel("x2")

centroids = Z[np.array([242, 528, 570, 590, 648, 667, 774, 891, 955])]
memberships = update_memberships(centroids, Z)

print(centroids)
print(centroids.shape)
while True:
    old_centroids = centroids
    centroids = update_centroids(memberships, Z)
    if np.alltrue(centroids == old_centroids):
        break

    old_memberships = memberships
    memberships = update_memberships(centroids, Z)
    if np.alltrue(memberships == old_memberships):
        plt.show()
        break

centroids = update_centroids(memberships, data_set)

plt.figure(figsize=(8, 8))
plot_current_state(centroids, memberships, data_set)
plt.show()
```
