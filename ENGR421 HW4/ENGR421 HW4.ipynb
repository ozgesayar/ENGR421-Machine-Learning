```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats

def safelog(x):
    return(np.log(x + 1e-100))
```

### Import data 

```python
dataset_train = np.genfromtxt('hw04_data_set_train.csv',delimiter=',')
dataset_test  = np.genfromtxt('hw04_data_set_test.csv',delimiter= ',')
```

```python
# get x and y values
x_train = dataset_train[:,0]
y_train = dataset_train[:,1]

x_test = dataset_test[:,0]
y_test = dataset_test[:,1]

# get number of classes and number of samples
K = np.max(y_train)
N = dataset_train.shape[0]
```

### Drawing Parameters

```python
point_colors = np.array(["red", "blue"])
minimum_value =  0.0
maximum_value = +2.0
```

### Regressogram 

```python
minimum_value = 0.0
maximum_value = 2.0

bin_width = 0.1
origin= 0.0

left_borders = np.arange(origin, maximum_value, bin_width)
right_borders = np.arange(origin + bin_width, maximum_value + bin_width, bin_width)
data_interval = np.arange(origin, np.max(x_train), 0.0005)
    
def regressogram(k):
    numerator = 0 
    denominator = 0 
    for i in range (len(x_train)):
        if(left_borders[k] < x_train[i] and x_train[i] <= right_borders[k]):
            numerator += y_train[i]
            denominator += 1
    return numerator/denominator

y_predicted = [regressogram(i) for i in range(left_borders.shape[0])]          

plt.figure(figsize=(13,5))
plt.scatter(x_train, y_train)
for b in range(len(left_borders)):
    plt.plot([left_borders[b], right_borders[b]], [y_predicted[b], y_predicted[b]], "k-")
for b in range(len(left_borders) - 1):
    plt.plot([right_borders[b], right_borders[b]], [y_predicted[b], y_predicted[b + 1]], "k-")
    
plt.figure(figsize=(13,5))
plt.scatter(x_test, y_test, color="red")
for b in range(len(left_borders)):
    plt.plot([left_borders[b], right_borders[b]], [y_predicted[b], y_predicted[b]], "k-")
for b in range(len(left_borders) - 1):
    plt.plot([right_borders[b], right_borders[b]], [y_predicted[b], y_predicted[b + 1]], "k-")
    

plt.title("Regressogram")
plt.show()
```

### Calculating RMSE for regressogram

```python
rmse = 0
for i in range (len(x_test)):
    for j in range (len(left_borders)):
        if(left_borders[j] < x_test[i] and x_test[i] <= right_borders[j]):
            rmse += (y_test[i] - y_predicted[int((x_test[i] - origin) / bin_width)]) ** 2
            
RMSE = np.sqrt(rmse / len(x_test))    
print("Regressogram => RMSE is", RMSE, " when h is", bin_width)
```

### Mean Smoother 

```python
data_interval = np.arange(origin, np.max(x_train), 0.001)

def mean_smoother(k):
    numerator = 0
    denominator = 0
    for i in range(len(x_train)):
        if(np.abs((data_interval[k]-x_train[i])/bin_width) <= 0.5):
            numerator +=  y_train[i]
            denominator += 1
    value = numerator / denominator
    return value

y_predicted = np.array([mean_smoother(i) for i in range(len(data_interval))])

plt.figure(figsize=(13, 5))
plt.plot(x_train, y_train, "b.", markersize=10, label='training')
plt.plot(data_interval, y_predicted, "k")
plt.show()

plt.figure(figsize=(13, 5))
plt.plot(x_train, y_train, "r.", markersize=10, label='test')
plt.plot(data_interval, y_predicted, "k")
plt.show()
```

### Calculating RMSE for Mean Smoother 

```python
rmse = 0
for i in range (len(x_test)):
    for j in range (len(left_borders)):
        if(left_borders[j] < x_test[i] and x_test[i] <= right_borders[j]):
            rmse += (y_test[i] - y_predicted[int((x_test[i] - origin) / bin_width)]) ** 2
            
RMSE = np.sqrt(rmse / len(x_test))    
print("Regressogram => RMSE is", RMSE, " when h is", bin_width)
```

### Kernel Smoother 

```python
bin_width = 0.02
def kernel_smoother(k):
    numerator = 0
    denominator = 0
    for i in range(len(x_train)):
        u = (data_interval[k] - x_train[i]) / bin_width
        g_x = 1/np.sqrt(2* np.pi) * np.exp(-1* u ** 2 / 2)
        denominator += g_x
        numerator += (g_x * y_train[i])
        
    val = numerator / denominator
    return val

y_predicted = np.array([kernel_smoother(i) for i in range(len(data_interval))])

plt.figure(figsize=(13, 5))
plt.plot(x_train, y_train, "b.", markersize=10, label='training')
plt.plot(data_interval, y_predicted, "k")
plt.show()

plt.figure(figsize=(13, 5))
plt.plot(x_train, y_train, "r.", markersize=10, label='test')
plt.plot(data_interval, y_predicted, "k")
plt.show()
```

### Calculating RMSE for Kernel Smoother 

```python
rmse = 0
for i in range (len(x_test)):
    for j in range (len(left_borders)):
        if(left_borders[j] < x_test[i] and x_test[i] <= right_borders[j]):
            rmse += (y_test[i] - y_predicted[int((x_test[i] - origin) / bin_width)]) ** 2
            
RMSE = np.sqrt(rmse / len(x_test))    
print("Regressogram => RMSE is", RMSE, " when h is", bin_width)
```
