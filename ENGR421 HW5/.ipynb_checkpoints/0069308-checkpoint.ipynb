```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def safelog2(x):
    if x == 0:
        return(0)
    else:
        return(np.log2(x))
```

### Import Data 

```python
# read data into memory
train_data = np.genfromtxt("hw05_data_set_train.csv", delimiter = ",")
test_data = np.genfromtxt("hw05_data_set_test.csv", delimiter = ",")

# get X and y values
x_train = train_data[:,0]
y_train = train_data[:,1]

x_test = test_data[:,0]
y_test = test_data[:,1]

# get number of classes, number of samples, and number of features
K_train = np.max(y_train).astype(int)
K_test = np.max(y_test)
N = x_train.shape[0]
```

## Decision Tree Regression Algorithm 

### Tree Inference

```python
# learning algorithm
def regression_algorithm(P):
    
    # create necessary data structures
    node_indices = {}
    is_terminal = {}
    need_split = {}

    node_splits = {} #If a node has P or fewer data points, do not split further
    node_means = {}

    # put all training instances into the root node
    node_indices[1] = np.array(range(train_data.shape[0]))
    is_terminal[1] = False
    need_split[1] = True
    
    while True:
        # find nodes that need splitting
        split_nodes = [key for key, value in need_split.items() if value == True]
        # check whether we reach all terminal nodes
        if len(split_nodes) == 0:
            break
        # find best split positions for all nodes
        for split_node in split_nodes:
            data_indices = node_indices[split_node]
            need_split[split_node] = False
            node_mean = np.mean(y_train[data_indices]) 
            if len(data_indices) <= P:
                is_terminal[split_node] = True
                node_means[split_node] = node_mean
            else:
                is_terminal[split_node] = False
                unique_values = np.sort(np.unique(x_train[data_indices]))
                split_positions = (unique_values[1:len(unique_values)] + \
                                    unique_values[0:(len(unique_values) - 1)]) / 2
                split_scores = np.repeat(0.0, len(split_positions))
                
                for pos in range(len(split_positions)):
                    left_indices = data_indices[x_train[data_indices] >= split_positions[pos]]
                    right_indices = data_indices[x_train[data_indices] < split_positions[pos]]
                    
                    score = 0.0
                    
                    for i in left_indices:
                        score += (y_train[i] - np.mean(y_train[left_indices])) ** 2
                    for i in right_indices:
                        score += (y_train[i] - np.mean(y_train[right_indices])) ** 2
                        
                    split_scores[pos] = score / len(data_indices)
                    
                node_splits[split_node] = split_positions[np.argmin(split_scores)]

                # create left node using the selected split
                left_indices = data_indices[x_train[data_indices] >= \
                               node_splits[split_node]]
                node_indices[2 * split_node] = left_indices
                is_terminal[2 * split_node] = False
                need_split[2 * split_node] = True

                # create right node using the selected split
                right_indices = data_indices[x_train[data_indices] <= \
                                node_splits[split_node]]
                node_indices[2 * split_node + 1] = right_indices
                is_terminal[2 * split_node + 1] = False
                need_split[2 * split_node + 1] = True

    return is_terminal, node_splits, node_means
```

```python
def predict(point, node_splits, is_terminal, node_means):
    index = 1
    while True:
        if is_terminal[index] == True:
            return node_means[index]
        else:
            if point > node_splits[index]:
                index = index * 2
            else:
                index = index * 2 + 1
    
```

### Figures 

```python
max_val = max(x_train)
origin = min(x_train)
data_points = np.linspace(origin - 0.1, max_val + 0.1, 1000)

is_terminal, node_splits, node_means = regression_algorithm(30)

y_predicted = []
for i in data_points:
    y_predicted = np.append(y_predicted, predict(i, node_splits, is_terminal, node_means))

plt.figure(figsize=(10, 6))
plt.plot(x_train, y_train, "b.", markersize=10, label='training')
plt.ylabel("Signal (milivolt)")
plt.xlabel("Time (sec)")
plt.legend(loc='upper right')
plt.plot(data_points, y_predicted, "k")
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x_test, y_test, "r.", markersize=10, label='test')
plt.ylabel("Signal (milivolt)")
plt.xlabel("Time (sec)")
plt.legend(loc='upper right')
plt.plot(data_points, y_predicted, "k")
plt.show()
```

### Calculating RMSE

```python
## RMSE Training
y_predicted_train = []
for i in x_train:
    y_predicted_train = np.append(y_predicted_train, predict(i, node_splits, is_terminal, node_means))
rmse_train = np.sqrt(np.mean((y_predicted_train-y_train)**2))

## RMSE Test
y_predicted_test = []
for i in x_test:
    y_predicted_test = np.append(y_predicted_test, predict(i, node_splits, is_terminal, node_means))
rmse_test = np.sqrt(np.mean((y_predicted_test-y_test)**2))

print("RMSE on training set is", rmse_train, "when P is 30")
print("RMSE on test set is", rmse_test, "when P is 30")
```

### Drawing RMSE for P = 10, 15, .... , 50

```python
P = [10, 15, 20, 25, 30, 35, 40, 45, 50]

## RMSE Training
rmse_train = []
for j in P:
    is_terminal, node_splits, node_means = regression_algorithm(j)
    y_predicted_train = []
    for i in x_train:
        y_predicted_train = np.append(y_predicted_train, predict(i, node_splits, is_terminal, node_means))
    rmse_train = np.append(rmse_train, np.sqrt(np.mean((y_predicted_train-y_train)**2)))
    
## RMSE Training
rmse_test = []
for j in P:
    is_terminal, node_splits, node_means = regression_algorithm(j)
    y_predicted_test = []
    for i in x_test:
        y_predicted_test = np.append(y_predicted_test, predict(i, node_splits, is_terminal, node_means))
    rmse_test = np.append(rmse_test, np.sqrt(np.mean((y_predicted_test-y_test)**2)))
    
plt.figure(figsize=(10, 6))
plt.plot(P, rmse_train, "b.-", markersize=10, label='training')
plt.plot(P, rmse_test, "r.-", markersize=10, label='test')
plt.xlabel("Pre-pruning size (P)")
plt.ylabel("RMSE")
plt.legend(loc='upper left')
plt.show()
```
